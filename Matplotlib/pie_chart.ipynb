# Pie Charts


```python
import matplotlib.pyplot as plt
```

- Market Share 

```python
list = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday']
shares=[1000,4320,4356,7445,4626,3636]

plt.pie(shares,labels=list, autopct='%1.2f%%')
plt.title('Market Share Pie Chart')
plt.show()
```

```python
list = ['A','B','C','D']
shares=[25,15,40,20]
color=['red','green','blue','yellow']
plt.pie(shares,labels=list, autopct='%1.2f%%')
plt.title('Market Share Pie Chart')
plt.show()
```

-  Population Distribution

```python
population=[45,30,25]
list= ['Children','Men','Women']

plt.pie(population,labels=list,autopct='%1.2f%%')
plt.title('Population')

plt.show()
```

-  Favorite Sports

```python
sports=['Cricket','Football','Tennis','BasketBall']
votes=[35,30,15,20]
c=['lightblue','red','purple','pink']

plt.pie(votes,labels=sports,colors=c,autopct='%1.2f%%')
plt.title('Sports Survey')
plt.show()
```

- Montly Expense Distribution

```python
pay = [1500,500,300,200,100]
categories = ['Rent','Food','Transportation','Entertainment','Miscellaneous']
c=['skyblue','peachpuff','mistyrose','beige','lightpink']

plt.pie(pay,labels=categories,colors=c,autopct='%1.1f%%',startangle=90)
plt.title('Montly Expense')
plt.show()
```
