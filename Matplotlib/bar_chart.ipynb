# Bar Chart

```python
import  matplotlib.pyplot as plt
```

- Vertical Simple Style

```python
x = ["Python","PHP","Java","TypeScript"]
y = [97,76,34,90]

plt.title("Programming Skills in 2024")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.bar(x,y)
plt.show()
```

- Horizontal Simple Style

```python
plt.barh(x,y)
```

- Vertical chart with different Colours

```python
x = ["Python","TypeScript","PHP","Java"]
y = [63,24,76,54]
c = ["red","green","blue","purple"]
plt.title("Programming Skills")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.bar(x,y,color=c)
plt.show()
```

- Horizontal chart with different Colours

```python
c = ["red","green","blue","purple"]
plt.barh(x,y,color=c)
```

# Vertical chart with Modification
- align
- color
- linewidth
-  width
- edgecolor

```python
x = ["Karachi","Lahore","Multan","Peshawar"]
y = [43,23,30,16]
c = ["red","grey","blue","yellow"]

plt.title("Weather of Pakistan in June")
plt.xlabel("Cities")
plt.ylabel("Temperature")

plt.bar(x,y,color = c,align='center',linewidth =3,width=0.4, edgecolor ="black")
plt.show()
```
