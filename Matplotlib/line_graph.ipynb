# Line Graph

```python
import matplotlib.pyplot as plt
```

- Simple Line Graph

```python
x = [1,2,3,5]
y = [43,23,65,34]
plt.title("Line Graph")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.plot(x,y)
plt.show()
```

- Plotting without line

```python
plt.plot(x,y,'p')
```

```python
s = [32,43,56,87,34,76,45,26]
plt.plot(s,color='red')
plt.show()
```

```python
# Marker Line Graph
plt.plot(s,marker='o')
```

-  Star Line Graph

```python
plt.plot(s,marker='*')
```

```python
# Line Plot with Grid

x = [10, 20, 30, 40, 50]
y = [100, 200, 150, 300, 250]

plt.title('Line Grid')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x,y,color='orange',marker='^')
plt.grid()
plt.show()
```

-  Grid with X-Axis

```python
plt.plot(x,y,color='orange',marker='^')
plt.grid(axis='x')
```

- Grid with Y-Axis

```python
plt.plot(x,y,color='orange',marker='^')
plt.grid(axis='y')
```

-  Custom Line Styles

```python
x = ['jan','feb','mar','apr','may','june']
y = [300,430,540,650,100,520]
plt.title('Sales of Six Months')
plt.xlabel('Months')
plt.ylabel('Sales')
plt.plot(x,y,color='purple', linestyle = 'dashed')
plt.show()
```

```python
plt.plot(x,y,color='purple', linestyle = 'dotted')
```

```python
plt.plot(x,y,color='purple', linestyle = '--')
```

```python
plt.plot(s, linewidth=10, color='blue')
```

- Two Lines Graph 

```python
x = [0, 1, 2, 3, 4]
y1 = [5, 10, 45, 20, 25]

x = [0, 1, 2, 3, 4]
y2 = [3, 54, 12, 18, 24]
plt.plot(x,y1,color='red',label = 'Line 1')
plt.plot(x,y2,color='blue',label = 'Line 2')
plt.legend()
plt.show()
```

- Three Lines Line Graph 

```python
x = [4,18,54,84,34]
y = [5, 10, 45, 20, 25]
z = [3, 54, 12, 18, 24]
plt.plot(x,color='red',label = 'Line 1')
plt.plot(y,color='blue',label = 'Line 2')
plt.plot(z,color='yellow',label = 'Line 3')
plt.legend()
plt.show()
```

- Legend keyword used

```python
x = [40,80,50,20,70]
y = [65,35,95,55,15]
z = [85,55,20,45,25]
plt.plot(x, color='r', label='PHP')
plt.plot(y, color='g', label='Python')
plt.plot(z, color='b', label='TypeScript')
plt.title("Programming Language Stats 2024")
plt.legend()
plt.show()
```
