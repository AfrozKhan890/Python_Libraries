# Histogram

```python
import matplotlib.pyplot as plt
```

- Basic

```python
data = [5, 7, 8, 9, 10, 10, 11, 13, 15, 16, 17, 18, 19, 20]

plt.hist(data,bins=10)
plt.xlabel('Data Values')
plt.ylabel('Frequency')
plt.show()
```

- Custom

```python
sales = [200, 150, 300, 400, 350, 500, 600, 550, 700, 750, 800, 850, 900]

plt.hist(sales,bins=range(100,1001,100),color='green',edgecolor='black')
plt.title("Sales Data Histogram")
plt.xlabel("Sales Range (in units)")
plt.ylabel("Frequency")
plt.show()
```

- Compare Two Datasets

```python
data1 = [60, 62, 64, 66, 68, 70, 72, 74, 76]
data2 = [65, 67, 69, 71, 73, 75, 77, 79, 81]

plt.hist(data1,bins=15,color='blue',alpha=0.6,edgecolor='black',label='Data1')
plt.hist(data2,bins=15,color='yellow',alpha=0.6,edgecolor='black',label='Data2')

plt.title("Histogram Comparison: Group A vs Group B")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
